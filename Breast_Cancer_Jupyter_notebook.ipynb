### Spliting the main dataframe in features dataframe X and target datafrane y

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

### Splitting the dataset into Training dataset (X_train, y_train) and Testing dataset (X_test, y_test)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

### Standardising the feature variables of better model Training

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### Training the SVM Classification model

```python
from sklearn.svm import SVC
classifier = SVC()
classifier.fit(X_train, y_train)
```

### Predicting the test result

```python
y_pred = classifier.predict(X_test)
```

### Creating the result Dataframe

```python
result_df = pd.DataFrame({"y_test" : y_test, "y_pred" : y_pred})
```

```python
result_df
```

### Evaluating the Model Performance

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5,3))
sns.heatmap(cm, annot=True)
print(cm)
```

## Evaluation :
#### A) Type 1 Error =4
#### B) Type 2 Error = 2

```python
print(classification_report(y_test, y_pred))
```

#### A) Model accuracy is 97%
#### B) F1 score is good, therefore data inbalance is not there
#### C) Let's try to improve the classification accuracy for Type 2 error

## Finding the best c-parameter, gamma-parameter and kernel using GridSearchCV

```python
param_grid = {'C' : [0.1, 1, 10, 100], 'gamma' : [1, 0.1, 0.01, 0.001], 'kernel' : ['rbf', 'linear']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=2)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid_pred = grid.predict(X_test)
```

### Recreating the confusion matrix

```python
cm = confusion_matrix(y_test, grid_pred)
plt.figure(figsize=(5,3))
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, grid_pred))
```

## Revaluation:
#### A) Model accuracy increased to 98%
#### B) Type 2 error reduced to 1
